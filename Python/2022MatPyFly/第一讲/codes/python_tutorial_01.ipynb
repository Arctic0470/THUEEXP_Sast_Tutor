#### 2.Python和 C 的主要区别
参考资料:[知乎:Python和C/C++的区别](https://www.zhihu.com/question/314756498?native.theme=1)

* C 是编译型语言，Python是解释型语言

编译型语言为了确保高级语言的代码在底层运行的效率，需要将整个程序作为一个整体进行编译。

解释型语言的每一行代码一般可独立执行，有利于进行实时、异步的调节，解释性语言难以实现静态类型检查

这种根本性的差异导致**Python语言是不需要像C/C++语言那样写一个入口函数(即int main)**,例如本次教程的第一个代码块。

* C 是弱类型、静态类型检查的，Python是强类型、动态类型检查的

简而言之就是C/C++中声明一个变量需要**确定类型**，不同的类型分配不同长度的字节，而Python则不需要，可以**自动扩展**

* Python是原生面向对象的语言，C 不是

C中需要使用一些编程手段可以实现面向对象编程(类的封装、继承、派生),而python中一切数据类型皆对象

* 一些细节区别

    * C语言的代码块用花括号{}标出，而Python则需要用缩进来区分不同层次的代码块。

    * C语言的各行用;隔开，而Python没有

    * 注释标记不同
    
    * ......

除了这些区别之外，Python的基本语法和数据类型和C语言是相近的

##### 3.3 字典(Dict)

特点:

* 字典中数据元素以key:value的形式给出

* key不可修改,决定了value的存储位置,value可修改

* key不重复,若重复则value为最后一个添加入字典的value;value可重复



```python
#3.3
#建立空字典
d = {}  #或者 d = dict{}
print(d)
#建立一个非空字典
d = {'中国':'China','美国':'America','日本':'Japan'}
print(d)
#通过key修改value
print(d['美国'])
d['中国'] = 'CN'
print(d)
#访问value
for x in d.values():
    print(x)
#访问key
for y in d.keys():
    print(y)
#同时访问key和value(以tuple的形式返回)
for (x,y) in d.items():
    print(x,'->',y)
#删除dict内的数据
del d['日本']
print(d)
#判断元素是否在dict内
B = '朝鲜' in d.keys()
print(B)
#清空dict
d.clear()
print(d)
```

##### 3.6 可迭代对象与迭代器(iterator)

* Python中内置的str、list、tuple、dict、set、file都是可迭代对象,均可以用for循环完成遍历,具体格式为"for x in",这种遍历称为迭代。

* 迭代器是一个可以记住遍历位置的对象,迭代器对象只能从集合的第一个元素开始向前访问,而不可用一个索引访问任意位置的元素

* Python中的迭代器在逻辑上类似于C语言中的链表

* 迭代器对象中有__iter__和__next__两种方法,而可迭代对象中只有__iter__,没有__next__

* 迭代器逐一取出数据元素放入内存,理论上可以表示无限大的数据流,而list、str、tuple这些不属于迭代器的可迭代对象则一次性将所有的元素从主存中取出加载到内存上


```python
#3.6
#迭代器的创建:可以用列表、元组、字符串等可迭代对象创建一个迭代器
list_iter = iter([1,2,3])#创建迭代器方法1
list_iter = [1,2,3].__iter__()  #创建迭代器方法2,用__iter__()返回一个迭代器
str_iter = iter('Tsinghua')
#调用__next__,观察输出变化
print(list_iter.__next__())
print(list_iter.__next__())
print(list_iter.__next__())
print(str_iter.__next__())
print(str_iter.__next__())
print(str_iter.__next__())
#print(list_iter[1])   #这行代码是错误的,迭代器无法根据索引进行访问
#判断是否是可迭代对象

from collections import Iterable  #需要调这个库才能执行下方的函数
print(isinstance(str_iter,Iterable))

#相比于C语言的链表,Python的迭代器对象遍历更为简易
for x in str_iter:
    print(x)
#留意代码的输出字符拼起来不是"Tsinghua"而是"nghua"
 
```

##### 3.7 生成器(generator)

相较于list、str、tuple等数据类型,迭代器只有在需要用到相应元素时才会将其加载到内存上,虽然不能使用索引,但是节约内存开销。生成器(generator)属于迭代器的概念范畴,调用生成器函数可以返回一个迭代器,理论上可以表示一个无限长的'数组'，这是list等数据结构所无法实现的。

特点:
* 生成器本质上是迭代器,只能通过__next__访问下一个元素,无法用索引访问
* 生成器保存了一套生成数值的算法,当需要新的值的时候,才返回

关于生成器的用法,见下方代码示例


```python
#3.7
L = [x for x in range(1,100)]#生成一个list
print(L)
gen = (x for x in range(1,100)) #生成一个generator
print(gen)

#用 yield关键字定义一个生成器函数[函数相关内容将在下一讲中介绍]
def Positive_Int():
    i = 0
    while i>=0:
        i = i+1
        yield i
g = Positive_Int()   #理论上来说,通过这个生成器能够产生一个无限长的正整数序列
#类似地,知道了一个序列[无论是数字还是字符]的递推关系之后,可以用生成器函数对其进行描述
for x in range(1,100,3):
    print(g.__next__())
```

##### 4.2 文件读写

Python的文件读写规则和C语言类似,而且相对更为简易,下面给出一些代码示例


```python
#4.2
filename = 'file1.txt'
f1 = open(filename,'w') #以写入方式打开文件,如果不存在文件则新建一个,若文件存在则清空
f1.write('Hello,World!')
f1.close()  #关闭文件

f2 = open('data.txt','r')
data = f2.read()  #一次性读取文件内全部内容
f2.close()
f2 = open('data.txt','r')
data = f2.readline() #读取一行内容
print(data)
#逐行读取,打印在屏幕上,并存储在列表中
Open = []
High = []
Low = []
Close = []
Volume = []
for x in f2.readlines():  #采用迭代的方法逐行读取文件信息
    print(x)
    open_temp,high_temp,low_temp,close_temp,volume_temp = x.split()  #采用Python内置的split函数,将对各行的字符串进行切片
    #split()默认以空格为分隔符,包括换行符'\n',其余输入情况可以尝试自行探索
    Open.append(float(open_temp))
    High.append(float(high_temp))
    Low.append(float(low_temp))
    Close.append(float(close_temp))
    Volume.append(int(volume_temp))
f2.close()

#调用绘图函数观察从文本文档中读取到的数据
import matplotlib.pyplot as plt
plt.figure(1)
plt.plot(Open)
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
#如果要在文件末尾追加新的信息,可以在open函数中'a'打开文件。
#上面的例子为文本文件的读写,若需要读写二进制文件,则在open函数中需要用'rb','wb','ab'等进行文件的打开

```
